# Įvadas į mašininį mokymąsi

## Greita istorija

- Mašininio mokymosi (MM, angl. Machine Learning) sąvoka pirmą kartą paminėta dar 1959 m., IBM tyrėjo Arthur Samuel --   vieno iš dirbtinio intelekto (DI, angl. Artificial Intelligence) srities pradininkų. MM dėl to dažnai vadinamas DI mokslo šaka, apimančia įvairios intelektinės veiklos, kaip mokymosi, suvokimo, loginio mąstymo ir problemų sprendimo aspektus.  

- Nuo to laiko, MM tapo savarankiška tarpdisciplinine sritimi. Joje daugiausia dėmesio skiriama mokymosi procesų automatizavimui panaudojant *tikimybių teorijos*, *statistikos*, *optimizavimo*, *informacijos teorijos* ir *algoritmų teorijos* rezultatus. 

- Dabar MM dažnai laikoma *informatikos* mokslų disciplina.

> [Machine Learning is the] field of study that gives computers the ability to learn without being explicitly programmed.
> 
> — **Arthur Samuel, 1959**

## Mokymasis iš duomenų

- MM kontekste mokymosi sąvoka sutapatinama su *mokymosi iš duomenų* (angl. learning from data) sąvoka.

- Mokymusi iš duomenų galima pavadinti procesą, kurio metu duomenyse esanti informacija "suvirškinama" ištraukiant esminius požymius bei, galiausiai, paverčiama rezultatais ar išvadomis. 


#### Pavyzdžiai (realaus mokymosi iš duomenų)

a) Gydytojai, norėdami diagnozuoti ligą, analizuoja pacientų simptomus, ligos istorijas, tyrimų rezultatus (duomenis). Įgaudami daugiau patirties, jie vis geriau atpažįsta simptomų ir tyrimų rezultatų dėsningumus, rodančius konkrečias ligas.

b) Vaikai klausosi juos supančios šnekamosios kalbos (duomenų). Laikui bėgant, bandydami mėgdžioti žodžius ir sakinius, jie išmoksta kalbos struktūrą ir taisykles.

<hr style="border: none; height: 2px; background-color: black;">

- Taigi, MM pastarąjį mokymąsi iš duomenų siekia aprašyti griežtai. 

- MM srityje dirbantys ekspertai siekia:
    - aiškiai apibrėžti mokymosi iš duomenų sąvoką;
    - kurti mokymosi modelius konkrečių uždavinių sprendimui;
    - kiekybiškai įvertinti mokymuisi reikalingus resursus;
    - vystyti efektyvius algoritmus, skirtus mokymosi automatizavimui;
    - kurti specializuotą programinę įrangą mokymosi modeliams įgyvendinti.

- MM sritis plėtojama įvairiai -- tiek taikant teorinę matematiką, siekiant apskaičiuoti įvairius įverčius ar užtikrinti teorines garantijas, tiek atliekant empirinius tyrimus, bandant palyginti algoritmų veikimus ar siekiant rasti tam tikras charakteristikas, tiek plėtojant praktinius programinės įrangos (angl. software) paketus.

> A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E.
> 
> — **Tom Mitchell, 1997**


## Tradicinis programavimas vs. mašininis mokymasis

- Taisyklių kūrimas:
    - **Tradicinis programavimas**: Taisyklės nustatomos rankiniu būdu, remiantis konkrečiais stebėjimais.
    - **Mašininis mokymasis**: MM algoritmai automatiškai išmoksta taisykles iš duomenų.
    
- Prisitaikymas:
    - **Tradicinis programavimas**: Laikui bėgant, taisykles reikia atnaujinti rankiniu būdu.
    - **Mašininis mokymasis**: MM modeliai gali automatiškai prisitaikyti prie pokyčių apmokant modelius su naujais duomenimis.

- Sudėtingumas:
    - **Tradicinis programavimas**: Sudėtingoms problemoms spręsti tradiciniai metodai dažnai reikalauja ilgos loginės sekos, komplikuotos kodo struktūros.
    - **Mašininis mokymasis**: MM supaprastina procesą, nes dėmesys sutelkiamas į duomenis, o sudėtingumą valdo mokymosi algoritmas, todėl kodas yra trumpesnis ir lengviau prižiūrimas.

## Mašininio mokymosi uždavinių tipai

Taikant tam tikrus skirstymo kriterijus MM srityje išskiriami keli uždavinių tipai:

1. **Prižiūrimas mokymasis (angl. supervised learning)**  
   Prižiūrimo mokymosi metu modelis yra mokomas naudojant duomenis, kurie turi nurodytas teisingas atsakymų žymes (angl. labels). Šis metodas naudojamas, kai turime didelį kiekį iš anksto sužymėtų duomenų, kurių pagrindu atliekamos prognozės.

   Prižiūrimas mokymasis toliau skirstomas į:

   1.1 **Klasifikavimo uždavinius (angl. classification)**

## Pagrindiniai mašininio mokymosi iššūkiai

Trumpai tariant, mokymosi procesas susideda iš modelio (algoritmo) parinkimo ir jo apmokymo ant tam tikrų duomenų, taigi, pagrindiniai ir paprasčiausi iššūkiai kurie gali kilti -- pasirinktas blogas algoritmas, arba prasti turimi duomenys.

1. **Nepakankamas mokymosi duomenų kiekis**  
   MM algoritmai efektyviai veikia tik tada, kai turi pakankamai duomenų; mažai duomenų lemia mažą modelio tikslumą.

2. **Nereprezentatyvūs duomenys**  
   Turėti vien daug duomenų neužtenka; šį iššūkį galime panagrinėti pavyzdžiu:


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

lifesat = pd.read_csv("https://github.com/ageron/data/raw/main/lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

model = LinearRegression()

model.fit(X, y)

fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(X, y, color='blue')
ax.plot(X, model.predict(X), color='black', label='Model predictions')
ax.set_xlim([23500, 65000])
ax.set_ylim([4, 9])
ax.set_xlabel('GDP per capita (USD)')
ax.set_ylabel('Life satisfaction')
ax.grid()
plt.show()
```

```python
missing_countries = {
    "South Africa": (20000, 4.2),
    "Colombia": (6000, 8.2),
    "Brazil": (18000, 7.8),
    "Mexico": (24000, 7.4),
    "Chile": (30000, 7.0),
    "Norway": (51000, 6.2),
    "Switzerland": (62000, 5.7),
    "Ireland": (81000, 5.2),
    "Luxembourg": (92000, 4.7),
}

missing_countries_df = pd.DataFrame.from_dict(
    missing_countries, 
    orient='index', 
    columns=['GDP per capita (USD)', 'Life satisfaction']
)
missing_countries_df = missing_countries_df.reset_index().rename(columns={'index': 'Country'})
```

```python
X_adt = missing_countries_df[["GDP per capita (USD)"]].values
y_adt = missing_countries_df[["Life satisfaction"]].values

X_full = np.concatenate((X, X_adt))
y_full = np.concatenate((y, y_adt))

model_new = LinearRegression()
model_new.fit(X_full, y_full)

fig, ax = plt.subplots(figsize=(10, 3))
ax.scatter(X, y, color='blue')
ax.scatter(X_adt, y_adt, color='red', marker='s')
ax.plot(X_full, model_new.predict(X_full), color='black', label='New model predictions')
ax.plot(X_full, model.predict(X_full), 'c:', label='Old model predictions')
ax.set_xlim([0, 115000])
ax.set_ylim([4, 9])
ax.set_xlabel('GDP per capita (USD)')
ax.set_ylabel('Life satisfaction')
ax.grid()
plt.legend()
plt.show()
```

Labai svarbu naudoti mokymosi duomenų aibę, kuri būtų reprezentatyvi tikram duomenų pasiskristymui. Tai dažnai yra sunkiau, nei atrodo: jei imtis yra per maža, atsiras imties triukšmas (t. y. nereprezentatyvūs duomenys dėl atsitiktinumo), tačiau net ir labai didelės imtys gali būti nereprezentatyvios, jei imties atrankos metodas yra blogas. Tai vadinama atrankos šališkumu (angl. sampling bias).

3. **Prasti duomenys**  
   Akivaizdu, kad jei mokymosi duomenyse yra daug klaidų, nukrypimų ir triukšmo (pvz., dėl prastos kokybės matavimų), modeliui bus sunkiau aptikti dėsningumus. Didelę dalį mašininio mokymosi proceso, veiksmų sekos (angl. pipeline) užima duomenų tvarkymas -- `NaN` reikšmių išvalymas, akivaizdžiai nuo normos nukrypusių verčių (angl. outliers) sutvarkymas, trūkstančių reikšmių užpildymas.  

2. **Persimokymas**  
   Šis iššūkis taip pat gali būti pavaizduotas pavyzdžiu:

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline

model = make_pipeline(PolynomialFeatures(degree=14), StandardScaler(), LinearRegression())
model.fit(X_full, y_full)

X_seq = np.linspace(X_full.min(), X_full.max(), 300).reshape(-1, 1)
y_seq = model.predict(X_seq)

fig, ax = plt.subplots(figsize=(10, 3))
ax.scatter(X, y, color='blue')
ax.scatter(X_adt, y_adt, color='red', marker='s')
ax.plot(X_seq, y_seq, color='black', label='Model predictions')
ax.set_xlim([0, 115000])
ax.set_ylim([4, 9])
ax.set_xlabel('GDP per capita (USD)')
ax.set_ylabel('Life satisfaction')
ax.grid()
plt.legend()
plt.show()
```

Sudėtingi modeliai gali aptikti nežymius duomenų dėsningumus, tačiau jei mokymosi duomenų aibė yra triukšminga arba per maža (dėl to atsiranda atrankos triukšmas), tikėtina, kad modelis aptiks dėsningumą pačiame triukšme. Akivaizdu, kad tokie modeliai prastai prognozuos naujus atvejus.

## Testavimas ir validavimas

Norint įvertinti modelio gebėjimą generalizuotis prie naujų duomenų, būtina atlikti modelio testavimą (angl. testing). Įsivaizduokite, kad mokote modelį spręsti tam tikrą problemą, tačiau norite būti tikri, kad jis veiks gerai ne tik su tais duomenimis, kuriuos turite dabar, bet ir su tais, kuriuos modelis pamatys ateityje. Kaip galite tai patikrinti? Vienas būdas tai padaryti yra <span style="visibility:hidden">padalinti turimus duomenis į dvi dalis: mokymo rinkinį ir testavimo rinkinį.</span>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split


np.random.seed(21)
X = np.random.normal(1, 10, 100)
y = 3 * X + 1 + np.random.normal(0, 10, 100)
X = X[:, np.newaxis]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
degrees = [i+1 for i in range(6)]
best_score = -np.inf
best_model = None

for degree in degrees:
    
    model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
    model.fit(X_train, y_train)

    y_test_pred = model.predict(X_test)
    score = model.score(X_test, y_test)
    if score > best_score:
        best_score = score
        best_model = model
    
    print(f"Degree {degree}: Accuracy score on test set = {score:.4f}")


X_seq = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_seq = best_model.predict(X_seq)

fig, ax = plt.subplots()
ax.scatter(X_train, y_train, color='blue', label='Train data')
ax.scatter(X_test, y_test, color='red', label='Test data')
ax.plot(X_seq, y_seq, color='black', label='Model predictions')
ax.grid()
ax.set_title(f"Best model is a degree {best_model.named_steps['polynomialfeatures'].degree} polynomial")
plt.legend()
plt.show()
```

Tačiau čia kyla problema – jeigu modelis bus per dažnai tikrinamas su ta pačia testavimo aibe, jis gali persimokyti ir šiems testavimo duomenims. Siekiant to išvengti, galima naudoti dar vieną duomenų aibę – validacijos (angl. validation) rinkinį. Šis rinkinys padės išsirinkti geriausią modelį ir jo parametrus. Galutinį modelio veikimą tikriname su testavimo duomenimis tik vieną kartą.

```python
from sklearn.model_selection import cross_val_score

degrees = [i+1 for i in range(5)]
best_score = -np.inf
best_model = None

for degree in degrees:
    
    model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
    scores = cross_val_score(model, X_train, y_train, cv=5)
    mean_score = scores.mean()

    if mean_score > best_score:
        best_score = mean_score
        best_model = model
    
    print(f"Degree {degree}: Mean validation accuracy = {mean_score:.4f}")


best_model.fit(X_train, y_train)
test_score = best_model.score(X_test, y_test)

X_seq = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_seq = best_model.predict(X_seq)

fig, ax = plt.subplots()
ax.scatter(X_train, y_train, color='blue', label='Train data')
ax.scatter(X_test, y_test, color='red', label='Test data')
ax.plot(X_seq, y_seq, color='black', label='Model predictions')
ax.grid()
ax.set_title(f"Best model is a degree {best_model.named_steps['polynomialfeatures'].degree} polynomial")
plt.legend()
plt.show()
```
