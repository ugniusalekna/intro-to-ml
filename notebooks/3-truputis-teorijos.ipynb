# Truputis mašininio mokymosi teorijos

Kadangi šiek tiek susipažinome su prižiūrimo MM eiga, tolimesnėms temoms pravers truputis statistinio mokymosi ir optimizavimo teorijos.

## Hipotezė, nuostolių funkcija, rizika

- Visų galimų požymių rinkinį žymėsime $\mathcal{X} \subseteq \mathbb{R}^k$. Šio rinkinio elementai yra požymių vektoriai, pažymėti mažosiomis raidėmis $x$ (arba $\textbf{x}$).

- Reikšmių rinkinys bus žymimas $\mathcal{Y}$ (baigtinė aibė klasifikavimo uždaviniuose, tolydi aibė regresijos uždaviniuose).

- $\mathcal{Z} = \mathcal{X} \times \mathcal{Y}$ žymėsime stebinių aibę.

- Bet kokia (mati) funkcija $h : \mathcal{X} \rightarrow \mathcal{Y}$ vadinama hipoteze ir gali būti naudojama prognozavimui.

- Visų galimų hipotezių rinkinį žymėsime $\mathcal{Y}^{\mathcal{X}}$. Paprastai nagrinėjame specifinius hipotezių rinkinius, o ne visą įmanomą hipotezių klasę (žymėsime $\mathcal{H} \subseteq \mathcal{Y}^{\mathcal{X}}$).

- Siekiant pamatuoti nuostolį, kurį patiriame prognozei taške $x$ naudodami hipotezę $h$, t.y. prognozuodami būsimą $y$ reikšmę lygią $h(x)$, nagrinėjamai hipotezių klasei $\mathcal{H}$ įvedama nuostolių funkcija $\ell : \mathcal{Z} \times \mathcal{H} \rightarrow [ 0; \infty )$. Reikšmė $l(z,h) = \ell((x,y),h)$ žymi nuostolį, kurį patirsime tikrąją stebinio reikšmę $y$ keisdami prognoze $h(x)$.

- $\ell(z,h)$ kiekybiškai įvertina nuostolį konkrečiai stebinio reikšmei, tačiau kadangi stebinys yra atsitiktinis dydis, hipotezė $h$ turėtų būti pakankamai tiksli visoms galimoms jo reikšmėms.

- Norint pasirinkti $h$ taip, kad lygybė $h(x) \approx y$ būtų kiek įmanoma tikslesnė visoms $x$ reikšmėms, įvedama rizikos funkcija $R(h) = \mathbb{E} [ \ell(z, h) ]$. Vidurkis imamas pagal $z$, funkcijos $R$ apibrėžimo sritis yra nagrinėjamos hipotezių klasės $\mathcal{H}$.

- $R(h)$ matuoja vidutinį hipotezės $h$ nuostolį fiksuotos nuostolių funkcijos $l$ atžvilgiu.

## Empirinės rizikos minimizavimo taisyklė

- Dėja, tačiau realiose situacijoje natūralu laikyti, kad atsitiktinio dydžio $z = (x,y)$ skirstinys yra nežinomas. Jeigu nežinome skirstinio, negalime apskaičiuoti jo vidurkio, t.y., dydžio $\mathbb{E} [ \ell(z, h) ]$. Taigi, pasirinkę nuostolių funkciją $\ell$ negalime suskaičiuoti ją atitinkančios rizikos – negalime ieškoti optimalios hipotezės (rizikos funkcijos minimumo taško) $\tilde{h} \in \text{arg} \min_{h \in \mathcal{H}} R(h)$.

- Siekiant rasti išeitį, apibrėžiama *empirinės rizikos funkcija*:

$$
R_{emp}(h) = \frac{1}{n} \sum_{i=1}^{n}{\ell(h, z_i) },
$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; kur $(z_1, ..., z_n)$ – mokymosi duomenų aibė.

- Remiantis **Didžiųjų skaičių dėsniu** (angl. the law of large numbers), artinant $n$ į begalybę, empirinės rizikos funkcija konverguoja į tikrąją riziką. Taigi, jei mokymosi duomenų aibės apimtis didelė, tai $R_{emp}(h) \approx R(h)$. Tuo paremta empirinės rizikos minimizavimo taisyklė – minimizuodami $R_{emp}$ gausime hipotezę $h$, kurios tikra rizika bus artima jos minimumui (infimumui).

## Optimizavimo pagrindai

- Optimizavimo uždavinys sprendžiamas taikant kone kiekvieną MM algoritmą (išimtis – tiesinės regresijos modelis, tačiau ir jo optimalūs parametrai gali būti ieškomi optimizavimo metodais).

- Optimizavimas atliekamas tuomet, kai ieškome optimalios hipotezės $h$ iš turimos hipotezių klasės $\mathcal{H}$. Pagrindinis optimizavimo uždavinys – rasti duotos funkcijos minimumo tašką, t.y. tašką, kuriame funkcija įgyja mažiausią reikšmę.

- Tegul $\mathcal{W} \in \mathbb{R}^k$, $f : \mathcal{W} \rightarrow \mathbb{R}$ – duota funkcija. Uždavinys – rasti funkcijos $f$ minimumo tašką $w^{*}$, t.y. tašką iš aibės $\text{arg} \min_{w \in \mathcal{W}} f(w)$.

### Sąvokos

- *Globalaus minimumo taškas* – toks taškas $w^* \in \mathcal{W}$, kad $f(w^*) \leq f(w)$ su visais $w \in \mathcal{W}$.

- *Lokalaus minimumo taškas* – toks taškas $w^* \in \mathcal{W}$, kad $f(w^*) \leq f(w)$ su visais $w$ iš tam tikros aplinkos, nelygios $\mathcal{W}$.

- *Funkcijos gradientas* – pirmų dalinių išvestinių vektorius. Gradientas taške $w$ paprastai žymimias $\nabla f(w)$. Gradiento išraiška taške $w$:

$$
\nabla f(w) := \left (  \frac{\partial f}{\partial w_1}(w), \frac{\partial f}{\partial w_2}(w), ..., \frac{\partial f}{\partial w_k}(w) \right )^T.
$$

Funkcijos gradientas rodo greičiausio funkcijos didėjimo taške $w$ kryptį, tuo tarpu, antigradientas (neigiama reikšmė) – didžiausio nuolydžio taške $w$ kryptį. Tuo iš tiesų nėra sunku įsitikinti:

Jeigu $u \in \mathcal{W} : || u || = 1$, tai kryptinė išvestinė $u \cdot \nabla f(w)$ rodo funkcijos $f$ didėjimo greitį $u$ kryptimi. Remiantis Koši-Švarco nelygybe skaliarinei sandaugai, galime įvertinti dydį:

$$
| u \cdot \nabla f(w) | \leq || u || \text{ } || \nabla f(w) || = || \nabla f(w) ||.
$$

Jei paimsime vienetinio ilgio vektorių gradiento kryptimi $ u_0 = \frac{\nabla f(w)}{|| \nabla f(w) ||}$, gausime, jog

$$
u_0 \cdot \nabla f(w) = \frac{\nabla f(w) \cdot \nabla f(w) }{|| \nabla f(w) ||} = || \nabla f(w) ||, 
$$

t.y. $\nabla f(w)$ kryptis tikrai sutampa su didžiausio didėjimo kryptimi!

### Gradientinio nusileidimo algoritmas 

<hr style="border: none; height: 2px; background-color: black;">

$$
\begin{aligned}
&\textbf{Require: } n \geq 1, w_0 \in \mathcal{W}, \epsilon > 0, \eta > 0 \cr
&\quad \text{Inicializuojame maks. iteracijų sk., pradinį artinį, tikslumą, žingsnio dydį} \cr
&\quad \text{Set } m \leftarrow 0 \cr
&\quad \text{Set } S \leftarrow 0 \cr
&\quad \textbf{while } m \leq n \textbf{ do} \cr
&\quad \quad m \leftarrow m + 1 \cr
&\quad \quad \text{Perskaičiuojame } w_m = w_{m-1} - \eta \nabla f(w_{m-1}) \cr
&\quad \quad S \leftarrow S + w_m \cr
&\quad \quad \textbf{if } \left| f(w_{m-1}) - f(w_m) \right| < \epsilon \textbf{ then} \cr
&\qquad \quad \text{Nutraukiame ciklą} \cr
&\quad \quad \textbf{end if} \cr
&\quad \textbf{end while} \cr
&\quad \text{return } \bar{w} = \frac{S}{m}
\end{aligned}
$$

<hr style="border: none; height: 2px; background-color: black;">

- Pradinis artinys $w_0$ gali būti parinktas lygiu atsitiktiniam apibrėžimo srities taškui.

- Žingsnio dydis $\eta$ dažniausiai parenkamas empiriškai. $\eta$ gali būti pastovus arba kintantis kiekviename žingsnyje, pagal tam tikrą taisyklę.

Išbandykime šį algoritmą patys:

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def function(w):
    return w**4 - 3*w**3 + 3*w

def gradient(w):
    return 4*w**3 - 9*w**2 + 3
```

```python
def gradient_descent(func, grad, initial_point, learning_rate=0.01, num_iterations=100, tol=1e-6):
    w = initial_point
    path = [w]
    
    for _ in range(num_iterations):
        new_w = w - learning_rate * grad(w)
        path.append(new_w)
        
        if abs(func(new_w) - func(w)) < tol:
            break
            
        w = new_w
    
    return np.array(path)
```

```python
def plot_gradient_descent(initial_point, learning_rate):
    path = gradient_descent(function, gradient, initial_point=initial_point, learning_rate=learning_rate)
    
    w = np.linspace(-1.5, 3, 100)
    f = function(w)
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.plot(w, f, label='Objective Function')
    ax.quiver(path[:-1], function(path[:-1]), path[1:] - path[:-1], function(path[1:]) - function(path[:-1]),
            scale_units='xy', angles='xy', scale=3, color='red')
    ax.plot(path, function(path), color='red', linestyle='--', label='Gradient Descent Path')
    
    final_point = path[-1]
    final_value = function(final_point)

    ax.annotate(f'Final Point\nw={final_point:.4f}\nf(w)={final_value:.4f}', 
                xy=(final_point, final_value), 
                xytext=(final_point + 0.1, final_value + 1),
            arrowprops=dict(facecolor='blue', shrink=0.05, headlength=5, headwidth=5, width=1),
                fontsize=10, color='black', backgroundcolor='white', bbox=dict(facecolor='white', alpha=0.8))

    ax.set_title(f"Gradient Descent (lr={learning_rate:.4f}, start={initial_point:.4f})")
    ax.set_xlabel('w')
    ax.set_ylabel('f(w)')
    ax.legend()
    ax.grid(True)
    
    plt.show()

# interact(plot_gradient_descent, 
#          initial_point=FloatSlider(value=1.0, min=-1.5, max=3, step=0.1, description='Initial Point'),
#          learning_rate=FloatSlider(value=0.1, min=0.01, max=0.5, step=0.01, description='Learning Rate', 
#                                    style={'description_width': '100px'}))

plot_gradient_descent(initial_point=1.0, learning_rate=0.1)
```

- Matome, jog algoritmas dažnai randa optimalų funkcijos minimumą, tačiau tai įvyksta ne visada – kokie pastebėjimai?

- Egzistuoja įvairių gradientinio nusileidimo modifikacijų, viena iš jų – momentinis metodas. Taikant šį metodą prie naujo taško pereinama naudojant taisyklę:

$$
w_{m+1} = w_m − \eta \nabla f(w_m) + \gamma \Delta w_m,
$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; kur $\Delta w_m = w_m − w_{m−1}$ ir $\gamma \in (0; 1]$ – fiksuota parametro reikšme.

```python
def gradient_descent_with_momentum(func, grad, initial_point, learning_rate=0.01, momentum=0.9, 
                                   num_iterations=100, tol=1e-6):
    w = initial_point
    w_diff = 0
    path = [w]
    
    for _ in range(num_iterations):
        new_w = w - learning_rate * grad(w) + momentum * w_diff
        w_diff = new_w - w
        path.append(new_w)
        
        if abs(func(new_w) - func(w)) < tol:
            break
            
        w = new_w
    
    return np.array(path)
```

```python
def plot_gradient_descent_with_momenum(initial_point, learning_rate, momentum):
    path = gradient_descent_with_momentum(function, gradient, 
                                          initial_point=initial_point, 
                                          learning_rate=learning_rate,
                                          momentum=momentum)
    
    w = np.linspace(-1.5, 3, 100)
    f = function(w)
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.plot(w, f, label='Objective Function')
    ax.quiver(path[:-1], function(path[:-1]), path[1:] - path[:-1], function(path[1:]) - function(path[:-1]),
            scale_units='xy', angles='xy', scale=3, color='red')
    ax.plot(path, function(path), color='red', linestyle='--', label='Gradient Descent Path')
    
    final_point = path[-1]
    final_value = function(final_point)

    ax.annotate(f'Final Point\nw={final_point:.4f}\nf(w)={final_value:.4f}', 
                xy=(final_point, final_value), 
                xytext=(final_point + 0.1, final_value + 1),
            arrowprops=dict(facecolor='blue', shrink=0.05, headlength=5, headwidth=5, width=1),
                fontsize=10, color='black', backgroundcolor='white', bbox=dict(facecolor='white', alpha=0.8))

    ax.set_title(f"Gradient Descent (lr={learning_rate:.4f}, start={initial_point:.4f}, momentum={momentum:.4f})")
    ax.set_xlabel('w')
    ax.set_ylabel('f(w)')
    ax.legend()
    ax.grid(True)
    
    plt.show()

# interact(plot_gradient_descent_with_momenum, 
#          initial_point=FloatSlider(value=1.0, min=-1.5, max=3, step=0.1, description='Initial Point'),
#          learning_rate=FloatSlider(value=0.1, min=0.001, max=0.01, step=0.001, description='Learning Rate', 
#                                    style={'description_width': '100px'}),
#          momentum=FloatSlider(value=0.9, min=0.0, max=1.0, step=0.01, description='Momentum'))

plot_gradient_descent_with_momenum(initial_point=-1.10, learning_rate=0.005, momentum=0.98)
```

### Teorinės gradientinio nusileidimo garantijos

- Aibė $A \in \mathbb{R}^k$ vadinama iškila, jei
$$
\forall w, v \in A \text{ } \forall t \in [0; 1] \text{ } tw + (1 − t )v \in A.
$$

Iškilos ir neiškilos aibės geometrinė interpretacija:

```python
import numpy as np
import matplotlib.pyplot as plt

convex_set = np.array([[1, 1], [3, 1], [2, 3], [1, 1]])
non_convex_set = np.array([[1, 1], [3, 1], [2, 2], [2, 4], [1, 1]])

def plot_sets():
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    ax[0].fill(convex_set[:, 0], convex_set[:, 1], color='skyblue', alpha=0.5, label='Convex Set')
    ax[0].plot(convex_set[:, 0], convex_set[:, 1], 'o-', color='blue')
    ax[0].set_title("Convex Set")
    ax[0].set_xlim(0, 4)
    ax[0].set_ylim(0, 5)
    ax[0].grid(True)
    
    for i in range(len(convex_set) - 1):
        for j in range(i + 1, len(convex_set)):
            ax[0].plot([convex_set[i, 0], convex_set[j, 0]], 
                       [convex_set[i, 1], convex_set[j, 1]], 
                       'r--')

    ax[1].fill(non_convex_set[:, 0], non_convex_set[:, 1], color='lightcoral', alpha=0.5, label='Non-Convex Set')
    ax[1].plot(non_convex_set[:, 0], non_convex_set[:, 1], 'o-', color='red')
    ax[1].set_title("Non-Convex Set")
    ax[1].set_xlim(0, 4)
    ax[1].set_ylim(0, 5)
    ax[1].grid(True)
    
    for i in range(len(non_convex_set) - 1):
        for j in range(i + 1, len(non_convex_set)):
            ax[1].plot([non_convex_set[i, 0], non_convex_set[j, 0]], 
                       [non_convex_set[i, 1], non_convex_set[j, 1]], 
                       'r--')

    plt.show()

plot_sets()
```

- Tegul $\mathcal{W} \in \mathbb{R}^k$ – iškila aibė. Funkcija $f : \mathcal{W} \rightarrow \mathbb{R}$ vadinama iškila, jei
$$
\forall w, v \in \mathcal{W} \text{ } \forall t \in [0; 1] \text{ } f( tw + (1 − t )v ) \leq tf(w) + (1 − t) f(v).
$$

Iškilos ir neiškilos funkcijos geometrinė interpretacija:

```python
import numpy as np
import matplotlib.pyplot as plt

def convex_function(x):
    return 0.5 * x**2

def non_convex_function(x):
    return x**3 - 3*x**2 + 2*x + 1

x = np.linspace(-1, 3, 400)
y_convex = convex_function(x)
y_non_convex = non_convex_function(x)

def plot_functions():
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    ax[0].plot(x, y_convex, label='Convex Function', color='blue')
    ax[0].fill_between(x, y_convex, y_convex.max(), color='skyblue', alpha=0.3)
    ax[0].set_title("Convex Function")
    ax[0].set_xlim(-1, 3)
    ax[0].set_ylim(-1, 5)
    ax[0].grid(True)
    
    x1, x2 = 0, 2
    ax[0].plot([x1, x2], [convex_function(x1), convex_function(x2)], 'r--')
    
    ax[1].plot(x, y_non_convex, label='Non-Convex Function', color='red')
    ax[1].fill_between(x, y_non_convex, y_non_convex.max(), color='lightcoral', alpha=0.3)
    ax[1].set_title("Non-Convex Function")
    ax[1].set_xlim(-1, 3)
    ax[1].set_ylim(-2, 5)
    ax[1].grid(True)
    
    x1, x2 = -0.5, 2.5
    ax[1].plot([x1, x2], [non_convex_function(x1), non_convex_function(x2)], 'r--')
    
    plt.show()

plot_functions()
```

- Teorinės gradientinio nusileidimo garantijos daugiausiai priklauso nuo minimizuojamos funkcijos savybių.
- Jei funkcija $f$ yra iškila, galima įrodyti, jog gradientinio nusileidimo algoritmas konverguos į *globalų* minimumą, kadangi tokios funkcijos kiekvienas lokalus minimumo taškas yra globalus minimumo taškas.
- Taikant kiek stipresnes prielaidas funkcijai $f$ bei apibrėžimo sričiai, galima apskaičiuoti tikslius konvergavimo greičio įverčius.
- Dėja, tačiau realiuose taikymo scenarijuose iškilos funkcijos pasitaiko retai – neiškilų funkcijų optimizavimas yra vienas sunkiausių uždavinių optimizavimo teorijoje. Šiuo metu praktikoje daugiausia taikomos įvairios gradientinio nusileidimo adaptacijos, pvz., momentinis gradientinis nusileidimas, kurį matėme aukščiau.

### Stochastinis gradientinis nusileidimas (angl. SGD)

- Pagrindinė gradientinio nusileidimo problema yra ta, kad kiekviename žingsnyje gradientai apskaičiuojami pagal visus funkcijos argumentus/parametrus/dėmenis; kuomet funkcija yra nesudėtinga ir vieno kintamojo, problemų tai nekelia, tačiau kuomet $w \in \mathbb{R}^k, k \gg 1$ arba funkcijoje dėmenų yra labai daug, jis veikia labai lėtai. 

- Stochastinis gradientinis nusileidimas kiekviename žingsnyje perskaičiuojant $w_{m}$ vietoje gradiento $\nabla f(w_{m-1})$ reikšmės naudoja atsitiktinę reikšmę $v_{m-1}$. Sekos $v_0, v_1, ... $ nariams keliama viena sąlyga – atsitiktinio dydžio $v_m$ vidurkis turi būti lygus funkcijos gradientui taške $w_m$.

- Dėl savo stochastinio pobūdžio šis algoritmas yra daug mažiau pastovus nei gradientinis nusileidimas: vietoje to, kad funkcijos reikšmės nuosekliai mažėtų link minimumo, mažėja tik jų vidurkis. Laikui bėgant reikšmės priartės prie minimumo, tačiau net ir pasiekus jį, vertės ir toliau svyruos, niekada nekonverguodamos į vieną tašką. Todėl, sustabdžius algoritmą, galutinės parametrų vertės bus geros, bet ne optimalios.

Stochastinio gradientinio nusileidimo veikimą paprasčiau pavaizduoti konkrečiu pavyzdžiu MM kontekste.
Sugrįžkime prie daugianarių regresijos pavyzdžio:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 100

X = 2 - 3 * np.random.normal(0, 1, n)
y = X - 2 * (X ** 2) + np.random.normal(-3, 10, n)

X = X[:, np.newaxis]

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.show()
```

- Daugianarių regresijos hipotezė taške $\textbf{x}$ apibrėžiama lygtimi:

$$
\hat{y} = \theta_0 + \theta_1 \textbf{x} + \theta_2 \textbf{x}^2 + ... + \theta_d \textbf{x}^d
$$

Pasinaudodami šia lygtimi galime apibrėžti netiktį $\ell$ taške $(\textbf{x}, y)$:

$$
\ell((\textbf{x}, y), h_{\boldsymbol{\theta}}) = \left( \theta_0 + \theta_1 \textbf{x} + \theta_2 \textbf{x}^2 + ... + \theta_d \textbf{x}^d - y \right)^2
$$

Matome, jog funkcija (funkcionalas) $\ell$ priklauso ne tik nuo vektoriaus $x$, tačiau ir nuo parametrų rinkinio $\boldsymbol{\theta}$. Taigi, $\ell = \ell(\textbf{x}, y; \boldsymbol{\theta})$.

Anksčiau užrašydami kvadratinės paklaidos lygtį, to net nežinodami apibrėžėme empirinę riziką:

$$
R_{emp}(\boldsymbol{\theta}) = \frac{1}{n} \sum_{i=1}^{n}{\ell((\textbf{x}, y), \boldsymbol{\theta}) } = \frac{1}{n} \sum_{i=1}^{n} \left( \theta_0 + \theta_1 \textbf{x} + \theta_2 \textbf{x}^2 + ... + \theta_d \textbf{x}^d - y \right)^2 = \text{MSE}(\textbf{x}, \boldsymbol{\theta})
$$


Empirinės rizikos minimizavimo taisyklė siūlo ieškoti optimalios hipotezės minimizuojant $R_{emp}$. Šį minimizavimo uždavinį galime išspręsti ne išreikštiniu būdu, tačiau SGD algoritmo pagalba. Tokiu atveju, optimizuojama funkcija yra $R_{emp}$, jos gradientas apibrėžiamas:

$$
\nabla R_{emp}(\boldsymbol{\theta}) = \frac{1}{n} \sum_{i=1}^{n}{\nabla \ell((\textbf{x}, y), \boldsymbol{\theta}) }
$$

*pastaba*: gradientai skaičiuojami ne pagal požymius $\textbf{x}$, tačiau pagal koeficientus $\boldsymbol{\theta}$.

Nesunku įsitikinti, jog

$$
\begin{align*}
\nabla \ell((\textbf{x}, y), \boldsymbol{\theta}) &= \nabla_{\boldsymbol{\theta}} \left( \left( \theta_0 + \theta_1 \textbf{x} + \theta_2 \textbf{x}^2 + \dots + \theta_d \textbf{x}^d - y \right)^2 \right) \cr &= 2 \left( \theta_0 + \theta_1 \textbf{x} + \theta_2 \textbf{x}^2 + \dots + \theta_d \textbf{x}^d - y \right) \begin{pmatrix} 1 \cr \textbf{x} \cr \textbf{x}^2 \cr \vdots \cr \textbf{x}^d \end{pmatrix} \end{align*}
$$

Arba užrašant trumpiau:


$$
\nabla \ell((\textbf{X}, y), \boldsymbol{\theta}) = 2 \left( \textbf{X} \boldsymbol{\theta} - y \right) \textbf{X}
$$


Nestochastinis gradientinis nusileidimas tinkamas, jei imtis $S$ maža. Jei $S$ didelė, tai $w_m = w_{m−1} − \eta \nabla R_{emp}(w_{m−1})$ skaičiavimo operacija gali pareikalauti daug mašininio laiko, nes $\nabla R_{emp}$ sumoje bus daug dėmenų.

Taikant SGN kiekviename žingsnyje galima apsiriboti tik vieno dėmens skaičiavimu:

Parinkdami atsitiktinį duomenų imties atvejį su vienoda tikimybe, apibrėžiame $v_m = \nabla \ell \left( \tilde{\textbf{x}}, \tilde{y}; \boldsymbol{\theta} \right)$, kur $(\tilde{\textbf{x}}, \tilde{y})$ – atsitiktinai parinkta apmokymo aibės elementų pora. Matome, jog taip apibrėžto atsitiktinio dydžio $v_m$ vidurkis tikrai lygus funkcijos gradientui taške $w_m$.

```python
import math

def stochastic_gradient_descent(X, y, learning_rate=0.01, num_iterations=100):
    n, k = X.shape
    d = 2 # polynomial degree
    num_parameters = int(math.factorial(k + d) / (math.factorial(k) * math.factorial(d)))

    theta = np.random.randn(num_parameters, 1) # random initial_point
    path = [theta]

    X_poly = np.c_[np.ones((n, 1)), X, X**2]

    for _ in range(num_iterations):

        random_index = np.random.randint(n)
        X_i = X_poly[random_index:random_index+1]
        y_i = y[random_index:random_index+1]
        
        gradients = 2 * X_i.T.dot(X_i.dot(theta) - y_i)

        theta = theta - learning_rate * gradients
        path.append(theta)
    
    return theta, path
```

```python
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X_standardized = (X - X_mean) / X_std

theta, path = stochastic_gradient_descent(X_standardized, y, learning_rate=0.01, num_iterations=250)
```

```python
import matplotlib.pyplot as plt

X_seq = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_seq_standardized = (X_seq - X_mean) / X_std

X_seq_poly = np.c_[np.ones((n, 1)), X_seq_standardized, X_seq_standardized**2]
y_pred_seq = X_seq_poly.dot(theta)

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(X, y)
ax.plot(X_seq, y_pred_seq, color='red', label='Polynomial Regression Curve')

ax.set_xlabel('X')
ax.set_ylabel('y')
ax.legend()
ax.grid()
plt.show()
```

Norėdami pamatyti stochastinę algoritmo prigimtį, galime pavaizduoti parametrų artėjimo į minimumo tašką "kelią". Kadangi mūsų rizikos funkcija priklauso nuo trijų parametrų rinkinio $(\theta_0, \theta_1, \theta_2)$, tokios funkcijos grafiškai pavaizduoti negalime, tačiau pasirinkdami dviejų parametrų poaibį rizikos funkciją pavaizduoti galime taip:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse

def plot_sgd_path_contour(path, X, y):
    
    theta_path = np.array(path).reshape(-1, 3)
        
    theta_1_seq = np.linspace(theta_path[:, 1].min() - 10, theta_path[:, 1].max() + 10, 100)
    theta_2_seq = np.linspace(theta_path[:, 2].min() - 10, theta_path[:, 2].max() + 10, 100)
    T1, T2 = np.meshgrid(theta_1_seq, theta_2_seq)

    X_poly = np.c_[np.ones((X.shape[0], 1)), X, X**2]
    
    Z = np.array([
        mse(y, X_poly.dot(np.array([[theta_path[-1, 0]], [t1], [t2]])))
        for t1, t2 in zip(np.ravel(T1), np.ravel(T2))
    ])
    
    Z = Z.reshape(T1.shape)

    plt.figure(figsize=(10, 7))
    plt.contourf(T1, T2, Z, levels=100, cmap="gray", alpha=0.7, zorder=1)
    plt.colorbar(label='Empirical Risk')

    plt.plot(theta_path[:, 1], theta_path[:, 2], marker='.', color='red', label='SGD Path', zorder=2)
    plt.scatter(theta_path[-1, 1], theta_path[-1, 2], marker='x', color='black', label='Final Theta', zorder=3)
    
    plt.xlabel(r'$\theta_1$')
    plt.ylabel(r'$\theta_2$')
    plt.title(r'SGD Path on Empirical Risk Contour')
    plt.legend()
    plt.grid()
    plt.show()


plot_sgd_path_contour(path, X_standardized, y)
```
